````python
def get_page(url):
    try:
        response=requests.get(url)
#       print (response.status_code)
        response.encoding='utf-8'
        html_content=response.text
#       print (html_content)
        url_pattern1 = re.compile('href="/czxx/index\.htm\?id=\d+&stationname=([\w\s]+)">')
        url_pattern2 = re.compile('<div class="axlcz(\d+)">')
        subway_station_menu =  url_pattern1.findall(html_content)
        subway_line_num     =  url_pattern2.findall(html_content)
        return subway_station_menu, subway_line_num
        
    except :
        return [],None
````

````python
def add_to_index(index,station,connect):
    if not connect : return
#    for entry in index[station]:
#        if  entry[0]==station :
    if  connect not in index[station] :
        index[station].append(connect) 
        return
#    index[station].append(connect)
           
````

````python
def add_page_to_index(stations,linename):
#    stations=content.split()   
#    subway_map[linenum]=[]
    for i,e in enumerate(stations):
        if linename not in subway_map[e]:
            subway_map[e].append(linename)
        if i==0 :
            add_to_index(index,e,stations[i+1])
            continue
        if i==len(stations)-1:
            add_to_index(index,e,stations[i-1])
            break
        add_to_index(index,e,stations[i-1])
        add_to_index(index,e,stations[i+1])
    return
````

````python
def creat_stationmap(subwaygroup):
    for item in subwaygroup:
        metroline,linenum=get_page(item)
        add_page_to_index(metroline,linenum[0])    
````

````python
def is_goal(desitination):
    def _wrap(current_path):
        return current_path[-1] == desitination
    return _wrap
````

````python
def sort_path(cmp_func):
    def _sort(pathes) :
        return sorted(pathes,key=cmp_func)
    return _sort
````

````python
def get_total_station(path):
    return len(path)
````

````python
def get_routeline(path):
    routeline=[]
    pre_s=path[0]
    for s in path[1:]:
        tmp=[val for val in subway_map[pre_s] if val in subway_map[s] ]             
  #      print (tmp)        
        pre_s=s
        if (not routeline):
            routeline.append(tmp)
        elif tmp != routeline[-1] :
            routeline.append(tmp)
        else:
            continue

    return routeline 
````

````python
def get_optimized(path):
    return len(path)+len(get_routeline(path))*3
````

````python
def search(graph, start, is_goal,search_strategy):
    pathes = [[start] ]
    seen = set()
    route=[]
    while pathes:
        path = pathes.pop(0)
        froniter = path[-1]        
        if froniter in seen: continue
        successors = graph[froniter]
        
        for city in successors: 
            if city in path: continue            
            new_path = path+[city]           
            pathes.append(new_path)       
            if is_goal(new_path):                             
                route=get_routeline(new_path)
                return new_path,route
#        print('len(pathes)={}'.format(pathes))
        seen.add(froniter)
        pathes = search_strategy(pathes)
````
